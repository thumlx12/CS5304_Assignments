```python
import pandas as pd
btc = pd.read_csv('btc.csv')
btc
```

```python
window_size = 10
means = [btc['Close'][i:i+window_size].mean() for i in range(len(btc)-window_size+1)]
variances = [btc['Close'][i:i+window_size].var() for i in range(len(btc)-window_size+1)]
print("MEANS:",means)
print("VARIANCES:",variances)
```
